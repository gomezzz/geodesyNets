# Loading results

```python
import pandas as pd
from gravann.input import csv_reader
import matplotlib.pyplot as plt

%matplotlib notebook
%load_ext autoreload
%autoreload 2
```

```python
FILEPATH = "./results"

df = csv_reader.read_result_csv(
    FILEPATH,
    include=["polyhedral-distance"],
    exclude=["checkpoint", "validation"]
)
df["Seed"] = df["Seed"].astype(int)
```

```python
df_rel = df[["Seed", "Ground Truth", "Target Sampler Domain", "Loss", "relRMSE@Altitude_0", "relRMSE@Altitude_1", "relRMSE@Altitude_2", "relRMSE@Altitude_3"]]

df_rel.columns = ["Seed", "Ground Truth", "Target Sampler Domain", "Loss", "0.001", "0.05", "1.0", "2.0"]

df_rel = df_rel.drop(columns=["Seed"])
df_rel
```

```python
df_rel.groupby(["Ground Truth", "Target Sampler Domain", "Loss"]).mean()
```

```python
df_tmp = df_rel[df_rel["Target Sampler Domain"] == "[0.0, 1.0]"].drop(columns=["Target Sampler Domain", "0.001", "0.05", "2.0"])

grouped = df_tmp.groupby(["Loss", "Ground Truth"]).mean().unstack(fill_value=0)

grouped.plot(kind="barh", ylabel="relRMSE", title="relRMSE", grid=True, figsize=(12, 4))
```

```python
df_tmp = df_rel[df_rel["Target Sampler Domain"] == "[1.0, 3.0]"].drop(columns=["Target Sampler Domain", "0.001", "0.05", "2.0"])

grouped = df_tmp.groupby(["Loss", "Ground Truth"]).mean().unstack(fill_value=0)

grouped.plot(kind="barh", ylabel="relRMSE", title="relRMSE", grid=True, figsize=(12, 4))
```
