# Loading results

```python
import pandas as pd
from gravann.io import csv_input_loader

%load_ext autoreload
%autoreload 2
```

```python
FILEPATH = "F:\\results"

df_noise = csv_input_loader.read_result_csv(FILEPATH, dir_substring="noise")

df_normal = csv_input_loader.read_result_csv(FILEPATH, dir_substring="polyhedral-distance")
```

```python
df_noise_rel = df_noise[["Seed", "Ground Truth", "Target Sampler Domain", "Noise Method", "relRMSE@Altitude_0", "relRMSE@Altitude_1", "relRMSE@Altitude_2", "relRMSE@Altitude_3", "relRMSE@Altitude_4", "relRMSE@Altitude_5", "relRMSE@Altitude_6"]]
df_noise_rel.columns = ["Seed", "Ground Truth", "Target Sampler Domain", "Noise Method", "0.01", "0.05", "0.1", "0.5", "1.0", "2.0", "5.0"]

df_noise_rel
```

```python
df_normal_rel = df_normal.loc[df_normal["Loss"] == "normalized_L1_loss"]
df_normal_rel = df_normal_rel[["Seed", "Ground Truth", "Target Sampler Domain", "Noise Method", "relRMSE@Altitude_0", "relRMSE@Altitude_1", "relRMSE@Altitude_2", "relRMSE@Altitude_3"]]
df_normal_rel["Noise Method"] = df_normal_rel["Noise Method"].fillna("None")

df_normal_rel.columns = ["Seed", "Ground Truth", "Target Sampler Domain", "Noise Method", "1e-3", "0.05", "1.0", "2.0"]

df_normal_rel = df_normal_rel.drop(columns=["1e-3"])

df_normal_rel
```

```python
total_df = df_noise_rel.append(df_normal_rel).fillna("None")
total_df = total_df.drop("Seed", axis=1)
total_df
```

```python
total_df.groupby(["Ground Truth", "Target Sampler Domain", "Noise Method"]).mean()
```
